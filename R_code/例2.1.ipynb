# 例2.1

```r
x<-c(3.4,1.8,4.6,2.3,3.1,5.5,0.7,3.0,2.6,4.3,2.1,1.1,6.1,4.8,3.8)
#生成数值向量x并赋予距消防站距离的数据
```

```r
y<-c(26.2,17.8,31.3,23.1,27.5,36.0,14.1,22.3,19.6,31.3,24.0,17.3,43.2,36.4,26.1)
#生成数值向量y并赋予火灾损失的数据
```

```r
mean(x)    #计算变量x的均值
sd(x)   #计算变量x的标准差
```

<img src='image/1.png' style="zoom:40%">

```r
plot(x,y,main="relationship",xlab="distance",ylab="loss")  #散点图
lm2.1<-lm(y~x)  #以y为因变量x为自变量建立回归方程
#并将结果赋给lm2.1其中默认回归方程是包含截距项的，如果是lm（y~x-1），则不包含截距项
abline(lm2.1,lwd=2, lty=1) # lty 参数用于设置线型，这里是实线
```

<img src='image/3.png' style="zoom:40%">

```r
summary(lm2.1)   #输出回归分析的结果
```

<img src='image/4.png' style="zoom:40%">

<img src='image/5.png' style="zoom:40%">

```r
anova(lm2.1)  #输出方差分析表
```

```r
cor(x,y,method = "pearson")  #计算x和y的相关系数
```

<img src='image/8.png' style="zoom:40%">

```r
cor.test(x,y,alternative = "two.sided",method = "pearson",conf.level = 0.95)  #检验相关系数的显著性代码
```

<img src='image/9.png' style="zoom:40%">

<img src='image/10.png' style="zoom:40%">

```r
e<-resid(lm2.1,digits=5) #将残差赋值给变量e，并保留小数点后5位
e  #在窗口中显示e的值
```

```r
# 绘制残差与拟合值的关系图
plot(fitted(lm2.1), e, 
     xlab = "x", ylab = "e",
     main = "Residual Plot for Simple Linear Regression")

abline(h = 0, lty = 2) # 添加一条水平参考线
```

<img src='image/11.png' style="zoom:40%">

<img src='image/12.png' style="zoom:40%">

```r
ZRE<-e/2.316  #计算标准化残差
SRE<-rstandard(lm2.1)  #计算学生化残差
ZRE
SRE
```

```r
new<-data.frame(x=3.5)   #输入新值3.5，此处必须以数据框的形式存储新点
ypred<-predict(lm2.1,new,interval = "prediction",level=0.95)
#计算预测值及预测区间并赋给ypred，此处level=0.95可以省略
yconf<-predict(lm2.1,new,interval = "confidence",level=0.95)
#计算预测值及置信区间并赋给yconf，此处level=0.95可以省略
ypred
yconf
```
